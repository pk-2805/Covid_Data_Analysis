```python
covid_df.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace=True, axis=1)
```

```python
covid_df.head(10)
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format='%Y-%m-%d')
```

```python
covid_df.head(5)
```

```python
# Create new column with Active Cases
covid_df['Active Cases'] = covid_df['Confirmed'] - covid_df['Cured'] - covid_df['Deaths']
covid_df.tail(5)
```

```python
# Pivot Table
statewise = pd.pivot_table(covid_df, values= ["Confirmed", "Deaths", "Cured"], index= "State/UnionTerritory", aggfunc= max)
```

```python
statewise.head(5)
```

```python
# find out the recovery rate
statewise['Recovery Rate'] = (statewise['Cured']/statewise['Confirmed'])*100
statewise.head(5)
```

```python
# find out the mortality rate
statewise['Mortality Rate'] = (statewise['Deaths']/statewise['Confirmed'])*100
statewise.head(5)
```

```python
# sort the values based on the confirmed cased and in descending order
statewise = statewise.sort_values(by = 'Confirmed', ascending= False)
```

```python
# Plot the pivot table using cmap
statewise.style.background_gradient(cmap= "cubehelix")
```

```python
# Top 10 states with active cases using gruopby
top_10_active_cases = covid_df.groupby(by= 'State/UnionTerritory').max()[['Active Cases', 'Date']].sort_values(by= ['Active Cases'], ascending= False).reset_index()
```

```python
fig = plt.figure(figsize=(16,9))
```

```python
plt.title('Top 10 states with most active cases in India', size = 12)
```

```python
# pass the axis using barplot function under seaborn lib
ax = sns.barplot(data = top_10_active_cases.iloc[:10], y = "Active Cases", x = "State/UnionTerritory", linewidth = 2, edgecolor = "red")
```

```python
# Top 10 states with active cases using gruopby
top_10_active_cases = covid_df.groupby(by= 'State/UnionTerritory').max()[['Active Cases', 'Date']].sort_values(by= ['Active Cases'], ascending= False).reset_index()
fig = plt.figure(figsize=(16,9))
plt.title('Top 10 states with most active cases in India', size = 12)
# pass the axis using barplot function defined under seaborn lib
ax = sns.barplot(data = top_10_active_cases.iloc[:10], y = "Active Cases", x = "State/UnionTerritory", linewidth = 1, edgecolor = "red")
plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show
```

```python
# Top states with highest mortality
top_10_deaths = covid_df.groupby(by = 'State/UnionTerritory').max()[["Deaths", 'Date']].sort_values(by= ["Deaths"], ascending= False).reset_index()
fig = plt.figure(figsize= (18, 5))
plt.title("Top 10 states with most deaths", size = 25)
ax = sns.barplot(data= top_10_deaths.iloc[:12], y= "Deaths", x= "State/UnionTerritory", linewidth = 1, edgecolor = "black")
plt.xlabel("States")
plt.ylabel("Total Death Cases")
plt.show
```

```python
# Remove rows where state is India
vaccine = vaccine_df[vaccine_df.State !="India"]
vaccine
```

```python
vaccine.rename(columns={'Total Individuals Vaccinated': "Total"}, inplace= True)
vaccine.head(5)
```

```python
# Most Vaccinated State
max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total', ascending= False)[:5]
max_vac
```

```python
fig = plt.figure(figsize= (10,5))
plt.title("Top 5 vaccinated states in India", size = 20)
x = sns.barplot(data= max_vac[:10], y= max_vac.Total, x=max_vac.index, linewidth = 2, edgecolor = "Black")
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```

```python
# Least Vaccinated State
least_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
least_vac = least_vac.sort_values('Total', ascending= True)[:5]
least_vac
```

```python
fig = plt.figure(figsize= (10,5))
plt.title("5 least vaccinated states in India", size = 20)
x = sns.barplot(data= least_vac[:10], y= least_vac.Total, x=least_vac.index, linewidth = 1, edgecolor = "Black")
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
